# Extracting the noises out of dataset

```python
df1 = df_NYA[((5500 < df_NYA["Open"]) & (df_NYA["Open"] < 6000) & (0 < df_NYA["Adj Close"]) & (df_NYA["Adj Close"] < 2000)) |
             ((0 < df_NYA["Open"]) & (df_NYA["Open"] < 1000) & (5800 < df_NYA["Adj Close"]) & (df_NYA["Adj Close"] < 6500))]
df1          # df1 refers to a dataframe showing the indices of noises
```

```python
df_NYA_1 = df_NYA.drop(df1.index)    # extracting the noises of the first plot by having their indeces 
df_NYA_1
```

```python
# plotting again to check whether it was correct 

plt.figure(figsize=(15,6))
plt.title("Open price vs. Adj Close price for NYA")
plt.scatter(df_NYA_1['Open'],df_NYA_1['Adj Close'],alpha=0.4)
plt.xlabel("Opening price")
plt.ylabel("Adj Close")
plt.grid()
```

<span style="color:green;font-weight:bold;font-size:12pt;font-style:italic;">&#10004; Two(2) noises are successfully removed</span>


```python
df2 = df_NYA[((5800 < df_NYA["High"]) & (df_NYA["High"] < 6500) & (0 < df_NYA["Adj Close"]) & (df_NYA["Adj Close"] < 2000)) |
             ((0 < df_NYA["High"]) & (df_NYA["High"] < 2000) & (5800 < df_NYA["Adj Close"]) & (df_NYA["Adj Close"] < 6500))]
df2          # df2 refers to a dataframe showing the indices of noises
```

<span style="color:blue;font-weight:bold;font-size:12pt;font-style:italic;"> those three(3) noises are recognized 

```python
df_NYA_2 = df_NYA.drop(df2.index)    # extracting the noises of the second plot by having their indeces 
```

```python
plt.figure(figsize=(15,6))
plt.title("Maximum price vs. Adj Close price for NYA")
plt.scatter(df_NYA_2['High'],df_NYA_2['Adj Close'],alpha=0.4)
plt.xlabel("Maximum price")
plt.ylabel("Adj Close")
plt.grid()
```

<span style="color:green;font-weight:bold;font-size:12pt;font-style:italic;">&#10004; Three(3) noises are successfully removed</span>

```python
df3 = df_NYA[((0 < df_NYA["Low"]) & (df_NYA["Low"] < 2000) & (5800 < df_NYA["Adj Close"]) & (df_NYA["Adj Close"] < 6500))]
df3          # df3 refers to a dataframe showing the indices of noises in Min price vs. Adj. close 
```

```python
df_NYA_3 = df_NYA.drop(df3.index)    # extracting the noises of the Third plot by having their indeces 
```

```python
plt.figure(figsize=(15,6))
plt.title("Minimum price vs. Adj Close price for NYA")
plt.scatter(df_NYA_3['Low'],df_NYA_3['Adj Close'],alpha=0.4)
plt.xlabel("Minimum price")
plt.ylabel("Adj Close")
plt.grid()
```

```python
df4 = df_NYA[((0 < df_NYA["Close"]) & (df_NYA["Close"] < 2000) & (5800 < df_NYA["Adj Close"]) & (df_NYA["Adj Close"] < 6500))]
df4          # df4 refers to a dataframe showing the indices of noises in Min price vs. Adj. close 
```

```python
df_NYA_4 = df_NYA.drop(df4.index)    # extracting the noises of the forth plot by having their indeces 
```

```python
# Combine index labels from all DataFrames
combined_index = df1.index.union(df2.index).union(df3.index).union(df4.index)

# Drop rows from df_NYA based on the combined index
filtered_data = df_NYA.drop(combined_index)

combined_index
```
